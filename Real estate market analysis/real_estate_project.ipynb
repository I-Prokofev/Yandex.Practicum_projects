#### Заполнение пропусков

Для пропусков в некоторых категориальных переменных можно предположить логичную замену:
* Если человек не указал, что квартира является апартаментами, вероятнее всего, это не апартаменты, а квартира: пропуски в столбце is_apartment правильно заменить на False.  
* Если человек не указал число балконов — скорее всего, их нет: пропуски в столбце balcony правильно заменить на 0.

Пропуски в переменных, для которых нет подходящего значения на замену и доля которых составляет менее 5% (locality_name и floors_total), мы можем отбросить.

Заполним пропуски в соответствующих столбцах:

```python
# заполним пропуски в столбце is_apartment
#print(len(df['is_apartment']))
#print(len(df[df['rooms']==0 & df['is_apartment'].isna()]))
df['is_apartment'] = df.loc[df['rooms']==0 & df['is_apartment'].isna(),'is_apartment'].fillna(True)
df['is_apartment'] = df['is_apartment'].fillna(False)

# заполним пропуски в столбце balcony
df['balcony'] = df['balcony'].fillna(0)

# отбросим пропуски в столбце locality_name и floors_total 
df = df.dropna(subset=['locality_name', 'floors_total'])

# проверим полученные данные
print(df.isna().mean())
```

Данные в столбцах ceiling_height, living_area, kitchen_area, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition являются количественными переменными. Пропуски в таких переменных следует заполнить характерным значением для выборки по населенному пункту.

Для этого сначала найдем число уникальных значений в столбце locality_name и определим нет ли в нем неявных дубликатов: 

```python
print('\nЧисло уникальных значений в столбце locality_name:')
display(df['locality_name'].value_counts(dropna=False).head(10))
display(df['locality_name'].value_counts(dropna=False).tail(10))
```

В столбце присутствуют неявные дубликаты, например: 'поселок Мурино' / 'Мурино', 'посёлок городского типа Рябово' / 'поселок городского типа Рябово', 'Кудрово' / 'деревня Кудрово' и т.д. 

Их необходимо обработать, для этого создадим функцию correct_locality_name, которая убирает из названия населенного пункта лишние данные, касающиеся его типа:

```python
# типы населенных пунктов, которые встречаются в названиях 
types_locality_name = ['садоводческое некоммерческое товарищество ',
                       'посёлок при железнодорожной станции ',
                       'посёлок городского типа ',
                       'поселок городского типа ',
                       'садовое товарищество ',
                       'коттеджный поселок ',
                       'городской посёлок ',
                       'городской поселок ',
                       'поселок станции ',
                       'посёлок станции ',
                       'посёлок ', 
                       'поселок ',
                       'деревня ',
                       'село ']

# функция correct_locality_name убирает из названия населенного пункта лишние данные:
def correct_locality_name(locality_name):
    for type_locality_name in types_locality_name:
        if type_locality_name in locality_name:
            return locality_name.replace(type_locality_name, '')
    return locality_name

# откорректируем столбец locality_name при помощи correct_locality_name
df['locality_name'] = df['locality_name'].apply(correct_locality_name)

#проверим получившиеся данные
print('\nЧисло уникальных значений в столбце locality_name:')
print(df['locality_name'].value_counts(dropna=False).head(10))
print(df['locality_name'].value_counts(dropna=False).tail(10))
```

Теперь заполним пропуски в некоторых столбцах с количественными переменными, которые в дальнейшем нам потребуются для исследований:
* ceiling_height - медианным значением; 
* living_area - средним значением отношения жилой площади к общей площади * общую площадь при группировке по количеству комнат;
* kitchen_area - средним значением отношения жилой площади к общей площади * общую площадь при группировке по количеству комнат;

```python
# заполним пропуски в ceiling_height
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

# Заполнение пропусков в living_area и kitchen_area:

# создадим в таблице столбцы living_ratio и kitchen_ratio, в которых запишем отношение 
# жилой площади к общей площади и площади кухни к общей площади, соотвественно:
df['living_ratio'] = df['living_area'] / df['total_area']
df['kitchen_ratio'] = df['kitchen_area'] / df['total_area']

# используя полученные столбцы, заполним пропуски в living_area и kitchen_area
df['living_area'] = df['living_area'].fillna(df.groupby(['rooms'])['living_ratio'].transform('mean')*df['total_area'])
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby(['rooms'])['kitchen_ratio'].transform('mean')*df['total_area'])

#проверим таблицу
print(df.isna().mean())
```

Пропуски в kitchen_area и days_exposition оставим как есть. 

Отбросим столбцы, где доля пропусков велика, и которые не потребуеются для дальнейшего анализа. Сохраним данные из airports_nearest, parks_around3000, parks_nearest, ponds_around3000 и ponds_nearest в переменной problem_stat, после чего удалим их из датасета:

### Изменение типов данных

Изменим типы данных в столбцах, чтобы они соответствовали предоставленной документации.

### Этаж квартиры

Добавим в таблицу столбец 'floor_category', в который запишем категорию этажа каждой квартиры.

Для этого создадим функцию find_floor_caterory, которая будет возвращать значения 'первый', 'последний' или 'другой' в зависимости от этажа квартиры:

```python
# функция возвращает значения 'первый', 'последний' или 'другой' в зависимости от этажа квартиры
def find_floor_caterory (row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
# добавим в таблицу столбец floor_category, используя find_floor_caterory
df['floor_category'] = df.apply(find_floor_caterory, axis=1)

display(df.tail(5))
```

Отношение жилой площади к общей площади, а также отношение площади кухни к общей площади мы уже считали на этапе предобработки данных. Обновим данные столбцы после заполнения пропусков и выведем датасет: </div>

```python
df['living_ratio'] = df['living_area'] / df['total_area']
df['kitchen_ratio'] = df['kitchen_area'] / df['total_area']
display(df.head(5))
```
